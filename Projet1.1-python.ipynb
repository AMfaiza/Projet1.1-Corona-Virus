#### Q6. Affichage des statistiques par pays en Algerie

#### Q7. Visualiser les pays les plus touchés

```python
# On va utiliser la librairie MATPLOTLIB pour la visualisation (on l'a importé en haut)
"""On prend par exemple 15 pays les plus touchés donc, premierement on va ordonner les cas confirmés en utilisant la méthode 
sort_values et faire les valeurs dans une liste """ 
# Ordonner les valeurs de la colonne 'Confirmed'
ordre_croissant=list(data4['Confirmed'].sort_values(na_position='first'))
# Afficher les 15 derniers éléments
data4['Confirmed']=ordre_croissant

plt.figure(figsize=(20,6))
# Scatter plot with day against tip
plt.scatter(data4['Country_Region'].iloc[-15:-1], ordre_croissant[-15:-1])
 
# Adding Title to the Plot
plt.title("Scatter Plot")
 
# Setting the X and Y labels
plt.xlabel('PAYS LES PLUS TOUCHES')
plt.ylabel('Nombre de cas confirmés')
 
plt.show()
```

```python
# La voici la liste des cas confirmés par ordre croissant 
ordre_croissant
```

**Remarque: Les valeurs des ordonnées = les valeurs de l'axe des y * 10^7**
